```python
def read_interesting_column(filename):
    data = np.loadtxt(filename, delimiter=',', skiprows=1, usecols=1)
    average_calculated = data[:-1].mean()
    average_given = data[-1]
    diff = average_calculated - average_given
    if abs(diff) > 1e-8:
        print("average of", filename, "does not match")
    #else:
    #    print("average of", filename, "does match")
    return data, average_calculated
```

```python
filename
```

```python
int(filename.split('/')[2])
```

```python
# Save averages in a Python dictionary
results = {}
for filename in filenames:
    experiment_num = int(filename.split('/')[2])
    data, average = read_interesting_column(filename)
    results[filename] = average
    print('average for experiment', experiment_num, ':', average)
```
